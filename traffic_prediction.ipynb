## Part 5 — Data Preprocessing 

## What does scaling do ?
- Scaing means changing numerical values so all columns/features are on a similar range/scale.Basically It transforms data to (Mean = 0, std dev =1)
- Formula : z = (x - mean) / std

## Why is it important?
- scaling is important because to ensure that all features are in equal range.
- for e.g. In this dataset, Speed_Limit can have values like 30, 100, or 195, while Driver_Alcohol contains only 0 or 1. Without scaling, the machine learning model may give more importance to Speed_Limit simply because its numbers are much larger, even alcohol consumption also affects accidents.scaling brings all our numerical values into similar range so model treats every feature equally.It helps us to improve performance and accuracy of the model

## Part 6 — Build Full Pipeline

### Combine preprocessing using ColumnTransformer
### Add your model to create a full Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder, StandardScaler
from sklearn.neighbors import KNeighborsClassifier

transformers = ColumnTransformer([("t1", StandardScaler(), [3,4,5,6,10,11]),
                                    ("t2", OneHotEncoder(), [0,1,8,9,12]),
                                    ("t3", OrdinalEncoder(), [2,7])
                                  ])
                                   
```

```python
X_train_trans= transformers.fit_transform(X_train)
X_test_trans= transformers.transform(X_test)
```

```python
X_train_trans = pd.DataFrame(X_train_trans)
X_test_trans = pd.DataFrame(X_test_trans)

print(X_train_trans.shape)
print(X_test_trans.shape)
```

```python
## Model Building
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train_trans,y_train)
```

```python
y_pred = knn.predict(X_test_trans)
print(y_pred.shape)
print(y_test.shape)
```

## How does a pipeline help prevent data leakage?
- A pipeline help prevent data leakage because it makes sure that every preprocessing step(Imputation, encoding,scaling) is learned only from Training data.
- First we train the model using model.fit(X_train, y_train), where the model learns patterns only from the training data
- After that, we use model.predict(X_test) to make predictions on new unseen data.
- Then, All preprocessing steps are learning only from training data.
- And then same transformations applied to the testing data.
